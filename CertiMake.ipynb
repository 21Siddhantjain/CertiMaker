```python
def PutText(img,text,pos,scale,thick):
    cv2.putText(img,text=text,org=pos, fontFace=cv2.FONT_HERSHEY_SIMPLEX,fontScale= scale,color=(0,0,0),thickness=thick,lineType=cv2.LINE_AA)

template = cv2.imread('Template2.jpg')
#template = cv2.cvtColor(template , cv2.COLOR_BGR2RGB)
img = cv2.imread('icon.png')
#img = cv2.cvtColor(img , cv2.COLOR_BGR2RGB)
img = cv2.resize(img,(0,0),img,0.1,0.1)

def PutImg():
    template[template.shape[0]-img.shape[0]-115:template.shape[0]-115,template.shape[1]-img.shape[1]-450:template.shape[1]-450] = img
    template[template.shape[0]-img.shape[0]-115:template.shape[0]-115,template.shape[1]-img.shape[1]-90:template.shape[1]-90] = img

PutText(template,'Name of Student',(340,291),1,2)
PutText(template,'Name Of Teacher 1',(210,528),0.5,1)
PutText(template,'Designation 1',(210,553),0.5,1)
PutText(template,'Name of Teacher 2',(570,528),0.5,1)
PutText(template,'Designation 2',(570,553),0.5,1)

while True:
    cv2.imshow('Certificate',template)
    if cv2.waitKey(1) & 0xFF == 27 :
        break
    if cv2.waitKey(1) & 0xFF == 100 :
        cv2.imwrite('UpdatedCerti.png',template)
        break
        
cv2.destroyAllWindows()

    
```

```python
import cv2
import numpy as np

drawing = False
ix,iy = -1,-1

def draw_rectangle(event,x,y,flags,param):
    global ix,iy,drawing,mode

    if event == cv2.EVENT_LBUTTONDOWN:
        drawing = True
        ix,iy = x,y

    elif event == cv2.EVENT_MOUSEMOVE:
        if drawing == True:
            cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
           

    elif event == cv2.EVENT_LBUTTONUP:
        drawing = False
        cv2.rectangle(img,(ix,iy),(x,y),(0,255,0),-1)
        
        
img = np.zeros((512,512,3), np.uint8)

cv2.namedWindow(winname='my_drawing')

cv2.setMouseCallback('my_drawing',draw_rectangle)

while True: 
    cv2.imshow('my_drawing',img)
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```
