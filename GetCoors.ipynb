```python
from tkinter import *
from tkinter import filedialog
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import cv2
from time import *
```

```python
window = Tk()

window.title("GUI")

window.geometry('512x512')
window.resizable(0,0)
pic = PhotoImage('icon.png')
window.call('wm', 'iconphoto', window._w, PhotoImage(file='icon.png'))

drawing = False 
ix,iy = -1,-1 
count = 0

def GetFile():
    secondary = Tk()
    filename = filedialog.askopenfilename()
    label = Label(secondary, text = "")
    label.pack()
    label.configure(text = filename)
    if (filename != ""):
        secondary.destroy()
        return filename
    
    secondary.mainloop()


    
def AddingCertificate():
    filename = GetFile()
    certi = cv2.imread(filename)
    certi = cv2.cvtColor(certi , cv2.COLOR_BGR2RGB)
    #cv2.imshow('Certificate',certi)

    def draw_rectangle(event,x,y,flags,param):
        global ix,iy,drawing,mode,count

        if event == cv2.EVENT_LBUTTONDOWN:
            drawing = True
            ix,iy = x,y

        elif event == cv2.EVENT_MOUSEMOVE:
            if drawing == True:
                cv2.rectangle(certi,(ix,iy),(x,y),(255,255,255),-1)


        elif event == cv2.EVENT_LBUTTONUP:
            drawing = False
            cv2.rectangle(certi,(ix,iy),(x,y),(255,255,255),-1)
            count += 1

    cv2.namedWindow(winname='Certificate')
    cv2.setMouseCallback('Certificate',draw_rectangle)

    global ix,iy,drawing,mode,count
    
    while True: 
        cv2.imshow('Certificate',certi)
        if cv2.waitKey(1) & 0xFF == 27:
            break
        if cv2.waitKey(1) & 0xFF == 100 :
            cv2.imwrite('UpdatedCerti.png',certi)
            break
        if count == 3 :
            count = 0
            break
            
    cv2.destroyAllWindows()
    

# initialdir =  "/", title = "Select A File", filetype =(("jpeg files","*.jpg"),("all files","*.*"))

WelcomeText = Label(window,text="Welcome",font=("ArialBold",50)).pack()

AddNames = Button(window , text="Add Names" , font=("ArialBold",15) , height = 2 , width = 30).pack()
AddCerti = Button(window , text="Add Certificate Template" , font=("ArialBold",15) , height = 2 , width = 30 , command = AddingCertificate).pack()
AddSign = Button(window , text="Add Signatures or images" , font=("ArialBold",15) , height = 2 , width = 30).pack()
AddText = Button(window , text="Add Text" , font=("ArialBold",15) , height = 2 , width = 30).pack()
DownloadImg = Button(window , text="Download Current Certificate" , font=("ArialBold",15) , height = 2 , width = 30).pack()
Exit = Button(window , text="Exit" , font=("ArialBold",15) , height = 2 , width = 30 ,command=window.destroy).pack()

window.mainloop()
```
